```python
import erdantic as erd
from erdantic.examples.pydantic import Party
```

```python
from dataclasses import dataclass

@dataclass
class Region:
    noc: str
    region: str
    notes: str

@dataclass
class Game_info:
    games: str
    year: int
    season: str
        
@dataclass
class Games:
    games: Game_info    
    city: str
        
@dataclass
class Event:
    event: str
    sport: str

@dataclass
class Athlete:
    athlete_id: int
    name: str
    sex: str
    age: float
    hight: float
    weight: float
    noc: Region 
    team: str
        
@dataclass
class Athlete_Event:
    athlete_id : Athlete
    games: Games
    event: Event
    medal: str
        
# Generate the ERD
erd.create(Athlete_Event, Athlete, Region, Games, Game_info, Event)
```
